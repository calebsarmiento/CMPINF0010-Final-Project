## Importing and Visualizing the Data

The data is imported and used in a dataframe which is then graphed.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('pools.csv')

pools_df = df[df['type'] == 'Pool']
pools_df = pools_df[pools_df['name'] != 'Highland - Children']

ax = plt.subplots(figsize=(10, 6))

plt.bar(pools_df['name'], pools_df['capacity'])
plt.xlabel('Pool Name')
plt.ylabel('Capacity')
plt.title('Pool Capacities')
plt.xticks(rotation=45, ha='right')
plt.show()
```

From this data, we can list the parks in descending order for capacity.

```python
neighborhoods_df = pools_df[['name','neighborhood','capacity']]

order_by_column = 'capacity'
neighborhoods_df = neighborhoods_df.sort_values(by=order_by_column, ascending=False)

neighborhoods_df = neighborhoods_df.reset_index(drop=True)
neighborhoods_df.index += 1

print('Pool Neighborhoods')
print(neighborhoods_df)
```
